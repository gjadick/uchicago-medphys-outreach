# show k-space planar waves, kx and ky

Distance from center = frequency\
Angle from horizontal =  tilt \
Complex portion = phase 

```python
def get_planar(idx_kx,idx_ky,Delta_kx):
    # given kx and ky indices, and Delta_kx
    # return the planar way wanted 
    # Assumed Delta_kx = Delta_ky
    # assume index for kx and ky starts at -.5 
    kx = -.5 + idx_kx*Delta_kx
    ky = -.5 + idx_ky*Delta_kx
    
    freq = np.sqrt(kx**2 + ky**2) #distance from center
    tilt = np.arctan(ky/kx)
    
    #print('freq: ',freq, '\ntilt:', tilt*180/np.pi)
    planar_range = np.linspace(-1,1,101)
    X,Y = np.meshgrid(planar_range,planar_range)
    #planar = np.sin((X*np.cos(tilt)+Y*(np.sin(tilt))))*freq 
    planar = np.sin(2*np.pi*(X*kx+Y*ky))*freq #kx and ky show rotation, phi is phase, freq is frequency
    return planar,freq,tilt

# also should include phase from complex portion??? 
# also wait complex part isn't ky???? ugh. 
# check over dissertation
```

```python
# this shows different points in k-space and the corresponding plane wave
%matplotlib
fig, (ax1,ax2) = pl.subplots(1,2)
for i in np.linspace(1,193,20):
    i = round(i)
    planar_1,frq,tlt = get_planar(i,i,1/193)
    
    #show planar wave
    ax1.imshow(planar_1,cmap = 'gray')

    # get fft of planar
    ft = np.fft.ifftshift(planar_1)
    ft = np.fft.fft2(ft)
    ft = np.fft.fftshift(ft)

    
    ax2.imshow(abs(ft))
    pl.title(f"{i},{frq:.2f},{tlt:.2f}")
    pl.pause(.5)
%matplotlib inline
```

```python
def DrawCircle(kspace,center,radius): #takes it in idx, not frequency values
    kspace_cropped = kspace
    kspacerange = np.linspace(1,193,193)
    X,Y = np.meshgrid(kspacerange,kspacerange)
    for j in range(len(X)):
        for k in range(len(Y)):
            r2 = (X[j,k]-95-center[0])**2 + (Y[j,k]-95-center[1])**2 #this is rsquared
            #have to fix this
            if r2 > radius**2:
                kspace_cropped[j][k] = 0
                print(f"cropped {round(r2)}, {round(radius**2)}")
            else:
                print(f"not {round(r2)}, {round(radius**2)}")
    return kspace_cropped
```
